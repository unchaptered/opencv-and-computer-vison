```python
import cv2
import time
```

```python
cap = cv2.VideoCapture('mysupervideo.mp4')

if cap.isOpened() == False:
    print('ERROR FILE NOT FOUND! or WRONG CODEC USED!')

while cap.isOpened():
    ret, frame = cap.read()
    
    if ret == True:
        
        # WRITER 100 FPS
        time.sleep(1/100)
        
        cv2.imshow('frame', frame)
        print('FRAME DETECTED!')
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
            
    else:
        break

cap.release()
cv2.destroyAllWindows()
print('END')
```
