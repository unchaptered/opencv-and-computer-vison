```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_image(filename):
    return cv2.imread('../../DATA/' + filename)
```

```python
def cvt_image(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
def display_image(image):
    figure = plt.figure(figsize=(6, 5))
    ax = figure.add_subplot(111)
    ax.imshow(image, cmap = 'gray')
```

```python
gorilla = load_image('gorilla.jpg')
gorilla.shape
```

```python
display_image(gorilla)
```

```python
histogram_values = cv2.calcHist(
    [gorilla],
    channels=[0],
    mask=None,
    histSize=[256],
    ranges=[0,256]
)
histogram_values.shape
```

```python
plt.plot(histogram_values)
```

```python
equalized_gorilla = cv2.equalizeHist(gorilla)
```

```python
histogram_equalized_values = cv2.calcHist(
    [equalized_gorilla],
    channels=[0],
    mask=None,
    histSize=[256],
    ranges=[0,256]
)
```

```python
plt.plot(histogram_equalized_values)
```

```python
def display_image_rgb(image):
    figure = plt.figure(figsize=(6, 5))
    ax = figure.add_subplot(111)
    ax.imshow(image)
```

```python
color_gorilla = load_image('gorilla.jpg')
```

```python
cvted_gorilla = cv2.cvtColor(color_gorilla, cv2.COLOR_BGR2RGB)
display_image_rgb(cvted_gorilla)
```

```python
hsv_gorilla = cv2.cvtColor(cvted_gorilla, cv2.COLOR_RGB2HSV)
```

```python
# Hue
# Saturation
# Value <-- Histogram Equalization Target
```

```python
hsv_gorilla [:,:,2] = cv2.equalizeHist(hsv_gorilla [:,:,2])
```

```python
equalized_gorilla = cv2.cvtColor(
    hsv_gorilla,
    cv2.COLOR_HSV2RGB
)
display_image_rgb(equalized_gorilla)
```
