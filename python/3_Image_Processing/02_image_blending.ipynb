```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image1 = cv2.imread('../../DATA/dog_backpack.png')
image2 = cv2.imread('../../DATA/watermark_no_copy.png')

print(type(image1), type(image2))
```

```python
rgb_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
rgb_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_image1)
```

```python
plt.imshow(rgb_image2)
```

```python
######################################
## BLENDING IMAGES OF THE SAME SIZE ##
######################################
```

```python
resized_image1 = cv2.resize(image1, (1200, 1200))
resized_image2 = cv2.resize(image2, (1200, 1200))
```

```python
blended_image = cv2.addWeighted(
    src1=resized_image1,
    alpha=0.8,
    src2=resized_image2,
    beta=0.2,
    gamma=0)
cvt_blended_image = cv2.cvtColor(blended_image, cv2.COLOR_BGR2RGB)
plt.imshow(cvt_blended_image)
```

```python
##################################################################
## OVERLAY SMALL IMAGE ON TOP OF A LARGER IMAGE ( NO BLENDING ) ##
##################################################################

######################################
## BLEND TOGETHER OF DIFRRENT SIZES ##
######################################
```

```python
image3 = cv2.cvtColor(
    cv2.imread('../../DATA/dog_backpack.png'),
    cv2.COLOR_BGR2RGB
)
image4 = cv2.resize(
    cv2.cvtColor(
        cv2.imread('../../DATA/watermark_no_copy.png'),
        cv2.COLOR_BGR2RGB
    ), (600, 600)
)
```

```python
print(image3.shape, image4.shape)
```

```python
large_image = image3
small_image = image4
```

```python
x_offset = 100
y_offset = 300
```

```python
x_end = x_offset + small_image.shape[1]
y_end = y_offset + small_image.shape[0]
```

```python
large_image[y_offset:y_end, x_offset:x_end] = small_image
```

```python
plt.imshow(large_image)
```
