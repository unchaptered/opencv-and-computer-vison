```python
######################################################################################
## Histogram is a visual representation of the distribution of a continuous feature ##
######################################################################################
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_image(filename):
    return cv2.imread('../../DATA/' + filename)
```

```python
def cvt_image(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
def display_image(image):
    figure = plt.figure(figsize=(6, 5))
    ax = figure.add_subplot(111)
    ax.imshow(image, cmap = 'gray')
```

```python
origin_horse_image = load_image('horse.jpg')        # ORIGINAL BGR OEPN-CV
cvt_horse_image = cvt_image(origin_horse_image)     # CONVERTED TO RGB FOR SHOW

origin_rainbow_image = load_image('rainbow.jpg')    # ORIGINAL BGR OEPN-CV
cvt_rainbow_image = cvt_image(origin_rainbow_image) # CONVERTED TO RGB FOR SHOW

origin_bricks_image = load_image('bricks.jpg')      # ORIGINAL BGR OEPN-CV
cvt_bricks_image = cvt_image(origin_bricks_image)   # CONVERTED TO RGB FOR SHOW
```

```python
display_image(cvt_horse_image)
display_image(cvt_rainbow_image)
display_image(cvt_bricks_image)
```

```python
# OEPN-CV BGR with cvt_bricks_image
histed_bricks_image = cv2.calcHist(
    [cvt_bricks_image],
    channels = [0],
    mask = None,
    histSize = [256],    # Max Value(255) + 1,
    ranges = [0, 256]    # Min(0) <= area < Max(256)
)
```

```python
plt.plot(histed_bricks_image)
```

```python
# OEPN-CV BGR with cvt_horse_image
histed_horse_image = cv2.calcHist(
    [cvt_horse_image],
    channels = [0],
    mask = None,
    histSize = [256],
    ranges = [0, 256]
)
```

```python
plt.plot(histed_horse_image)
```

```python
#################
## blue bricks ##
#################
```

```python
target_image = cvt_bricks_image
target_color_list = ('b', 'g', 'r')

for idx, target_color in enumerate(target_color_list):
    histr = cv2.calcHist(
        [target_image],
        [idx],
        None,
        [256],
        [0, 256]
    )
    plt.plot(histr, color = target_color)
```

```python
target_image = cvt_horse_image
target_color_list = ('b', 'g', 'r')

for idx, target_color in enumerate(target_color_list):
    histr = cv2.calcHist(
        [target_image],
        [idx],
        None,
        [256],
        [0, 256]
    )
    plt.plot(histr, color = target_color)
    plt.xlim([0, 50])
    plt.ylim([0, 500000])
plt.title('HISTOGRAM FOR HORSE')
```
