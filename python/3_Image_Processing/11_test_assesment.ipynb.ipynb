**TASK:Apply a binary threshold onto the image.**

**TASK: Open the giaraffes.jpg file from the DATA folder and convert its colorspace to  HSV and display the image.**

```python
giraffes = cv2.imread('../../DATA/giraffes.jpg')
cvted_giraffes = cv2.cvtColor(giraffes, cv2.COLOR_BGR2RGB)
plt.imshow(cvted_giraffes)
```

```python
hsv_giraffes = cv2.cvtColor(cvted_giraffes, cv2.COLOR_RGB2HSV)
plt.imshow(hsv_giraffes)
```

**TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)**

```python
kernel = np.ones((4, 4), np.float32) / 10
```

```python
kernel
```

```python
convolutioned_giraffe = cv2.filter2D(
    src=cvted_giraffes,
    ddepth=-1,
    kernel=kernel,
)
plt.imshow(convolutioned_giraffe)
```

**TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.**

```python
giraffes = cv2.imread('../../DATA/giraffes.jpg', 0)
```

```python
sobel_x_image = cv2.Sobel(
    giraffes,
    cv2.CV_64F,
    1,
    0,
    ksize = 5
)
plt.imshow(sobel_x_image, cmap='gray')
```

**TASK: Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image. Pay careful attention to the ordering of the channels.**

```python
target_giraffes = cv2.imread('../../DATA/giraffes.jpg')
target_giraffes.shape
```

```python
target_color_list = ('b', 'g', 'r')

for idx, target_color in enumerate(target_color_list):
    histr = cv2.calcHist(
        [target_giraffes],
        [idx],
        None,
        [256],
        [0, 256]
    )
    plt.plot(histr, color = target_color)
```
