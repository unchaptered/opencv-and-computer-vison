```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('../../DATA/flat_chessboard.png')
plt.imshow(flat_chess)

####################################################################################
## For grid detection, grids has to looks like chess boards or be formalized grid ##
####################################################################################
```

```python
found, corners = cv2.findChessboardCorners(
    flat_chess,
    (7, 7)     # if you type (8, 8), this method couldn't find last grid
)
print(found)
```

```python
cv2.drawChessboardCorners(
    flat_chess,
    (7, 7),
    corners,
    found
)
plt.imshow(flat_chess)
```
