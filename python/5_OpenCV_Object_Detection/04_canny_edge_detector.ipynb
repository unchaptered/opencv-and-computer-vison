```python
#################################################################################################################
## Canny Edge Detector has some process                                                                        ##
##  1. Smoothing : Gausian Filter to remove noise                                                              ##
##    p.s. 배경 노이즈, 배경 피처, 전경 피처가 많은 이미지에서 엣지 검출을 사용하면 모서리가 너무 많이 감지됨    ##
##         가우시안 필터를 적용하여 이미지를 스무딩하면 이미지의 강도 증감률을 찾을 수 있음                      ##
##  2. Apply non-maximum suppression to get rid of supriousf response to edge detection                        ##
##  3. Apply double threshold to determine potential edges                                                     ##
##  4. Track edge by hysteresis : Finalize the detection of edges by suppressing all the other edges           ##
##       that are weak and not connected to strong edges                                                       ##
#################################################################################################################
```

```python
########################################################################################
## For high resolution images where you only want **GENERAL EDGES**,                  ##
##   it usally a good idea to apply a custom blur before applying the Canny Algorihtm ##
########################################################################################
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#################################################
## It doesn't matter if the order is mixed up. ##
#################################################

image = cv2.imread('../../DATA/sammy_face.jpg')
plt.imshow(image)
```

```python
edges = cv2.Canny(
    image = image,
    threshold1 = 127,
    threshold2 = 127
)
plt.imshow(edges)
##########################################################################
## 1. Blur on the image to get rid of details as general edge detection ##
## 2. Change threshold1 and 2                                           ##
##########################################################################
```

```python
#############################
## Change threshold1 and 2 ##
#############################

edges = cv2.Canny(
    image = image,
    threshold1 = 0,
    threshold2 = 255
)
plt.imshow(edges)
```

```python
median_value = np.median(image)
print(median_value)
```

```python
################################################################################
## Lower threshold to either 0 or 70% of the median value wichever is greater ##
################################################################################
lower = int(
    max(0, 0.7 * median_value)
)


############################################################################################
## Upper threshold to either 130 of the median value or the max 255, whichever is smaller ##
############################################################################################
upper = int(
    min(255, 1.3 * median_value)
)

print(lower, upper)
```

```python
edges = cv2.Canny(
    image = image,
    threshold1 = lower,
    threshold2 = upper
)
plt.imshow(edges)
```

```python
edges = cv2.Canny(
    image = image,
    threshold1 = lower,
    threshold2 = upper + 50
)
plt.imshow(edges)
```

```python
edges = cv2.Canny(
    image = image,
    threshold1 = lower,
    threshold2 = upper + 100
)
plt.imshow(edges)
```

```python
edges = cv2.Canny(
    image = image,
    threshold1 = lower,
    threshold2 = upper + 150
)
plt.imshow(edges)
```
