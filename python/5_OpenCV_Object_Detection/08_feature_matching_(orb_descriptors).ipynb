```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(image, cmap = 'gray'):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(image, cmap)
```

```python
####################################################################
## Feature Matching                                               ##
##  1. Brute-force matching with ORB Descriptors                  ##
##  2. Brute-force matching with SIFT Descriptors and Ratio Test  ##
##  3. FLANN based Mathcer                                        ##
####################################################################
```

```python
reeses = cv2.imread('../../DATA/reeses_puffs.png', 0)
display(reeses)
```

```python
cereals = cv2.imread('../../DATA/many_cereals.jpg', 0)
display(cereals)
```

```python
###############################################
## Brute-force matching with ORB Descriptors ##
###############################################
```

```python
orb = cv2.ORB_create()
```

```python
keypoint_1, description_1 = orb.detectAndCompute(reeses, None)
keypoint_2, description_2 = orb.detectAndCompute(cereals, None)
```

```python
bruce_force = cv2.BFMatcher(
    cv2.NORM_HAMMING,
    crossCheck = True
)
```

```python
matches = bruce_force.match(description_1, description_2)

single_match = matches[0]
print(single_match.distance)
```

```python
matches = sorted(matches, key=lambda X:X.distance)
```

```python
reeses_matches = cv2.drawMatches(
    reeses,
    keypoint_1,
    cereals,
    keypoint_2,
    matches[:25],
    None
)
```
