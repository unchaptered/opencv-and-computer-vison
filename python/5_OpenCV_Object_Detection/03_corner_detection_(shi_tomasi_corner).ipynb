```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
real_chess = cv2.imread('../../DATA/real_chessboard.jpg')
flat_chess = cv2.imread('../../DATA/flat_chessboard.png')

cvt_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)
cvt_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)

gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2GRAY)
gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2GRAY)
```

```python
################################################
## goodFeaturesToTrack                        ##
##  1. Pros : efficient                       ##
##  2. Cons : hand draw corner point yourself ##
################################################

corner = cv2.goodFeaturesToTrack(
    gray_flat_chess,
    64,
    0.01, # 최소 고유값 또는 본질적인 헤리스 함수
    10
)
print(type(corner[0][0][0]))
```

```python
corner = np.int0(corner)
print(type(corner[0][0][0]))
```

```python
for i in corner:
    # i = [[308 308]]
    # 308, 308 = [308.308]
    x, y = i.ravel()
    cv2.circle(
        cvt_flat_chess,
        (x, y),
        3,
        (255, 0, 0),
        thickness = -1
    )
plt.imshow(cvt_flat_chess)
```

```python
################################################
## goodFeaturesToTrack                        ##
##  1. Pros : efficient                       ##
##  2. Cons : hand draw corner point yourself ##
################################################

corners = cv2.goodFeaturesToTrack(
    gray_real_chess,
    80,
    0.01,
    10
)
print(type(corners[0][0][0]))
```

```python
corners = np.int0(corners)
print(type(corners[0][0][0]))
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(cvt_real_chess, (x, y), 3, (255, 0, 0), thickness = -1)

plt.imshow(cvt_real_chess)
```
