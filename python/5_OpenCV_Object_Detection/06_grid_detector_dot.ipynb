```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dots = cv2.imread('../../DATA/dot_grid.png')
plt.imshow(dots)
```

```python
found, corners = cv2.findCirclesGrid(dots, (10, 10), cv2.CALIB_CB_SYMMETRIC_GRID)
print(found)
```

```python
cv2.drawChessboardCorners(dots, (10, 10), corners, found)
plt.imshow(dots)
```
