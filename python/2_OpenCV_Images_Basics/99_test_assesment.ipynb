#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('../../DATA/dog_backpack.jpg')
converted_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(converted_image)
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
rotated_image = cv2.flip(converted_image, 0)
plt.imshow(rotated_image)
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
drawed_image = converted_image.copy()
cv2.rectangle(drawed_image, (200, 370), (600, 720), (255, 0, 0), 10)
plt.imshow(drawed_image)
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
triangled_image = converted_image.copy()
points = np.array([[400, 370], [200, 720], [600,720]], np.int32)
cv2.polylines(triangled_image,
              [points],
              True, (0, 0, 255), 10)
plt.imshow(triangled_image)
```
