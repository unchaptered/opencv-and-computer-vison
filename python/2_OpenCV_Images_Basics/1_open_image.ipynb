```python
# Resize Image
```

```python
pre_resized_image = cv2.imread('../../DATA/00-puppy.jpg')
aranged_resized_image = cv2.cvtColor(pre_resized_image, cv2.COLOR_BGR2RGB)
aft_resized_image = cv2.resize(aranged_resized_image, (1000, 400))
type(aft_resized_image)
```

```python
plt.imshow(aft_resized_image)
```

```python
# Resize Image with Ratio : Ratio
```

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
aft_resized_image_2 = cv2.resize(aranged_resized_image, (0, 0), aranged_resized_image, w_ratio, h_ratio)
```

```python
print(pre_resized_image.shape, aft_resized_image_2.shape)
```

```python
plt.imshow(aft_resized_image_2)
```

```python
# Flip
```

```python
pre_flipped_image = cv2.imread('../../DATA/00-puppy.jpg')
aranged_flipped_image = cv2.cvtColor(pre_flipped_image, cv2.COLOR_BGR2RGB)
aft_flipped_image = cv2.flip(aranged_flipped_image, -1)
plt.imshow(aft_flipped_image)
```

```python
# Save
```

```python
cv2.imwrite('totally_new.jpg', aft_flipped_image)
```

```python
# Show file with matplotlib
```

```python
fig = plt.figure(figsize = (2, 2))
ax = fig.add_subplot(111) # Single Image
ax.imshow(aft_flipped_image)
```
